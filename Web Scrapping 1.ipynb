```python
# https://www.sharesansar.com/today-share-price
    
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.sharesansar.com/today-share-price"
r=requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
soup
```

```python
table = soup.find("table")
table
```

```python
headers = table.find_all("th")
# headers
titles=[]
for i in headers:
     title = i.text
     titles.append(title)
df = pd.DataFrame(columns = titles)
df
```

```python
rows = table.find_all('tr')

for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text.strip() for tr in data]
#     print(row)
    l= len(df)
    df.loc[l]=row
df.to_csv("stockMarket.csv")
df
```
